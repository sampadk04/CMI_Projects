# 1. Loading

```python
# extract the paths to all the .txt files in the `data/raw` folder
raw_folder = os.path.join('data', 'raw')

corpus_files = glob.glob(os.path.join(raw_folder, '*.txt'))
```

# 2. Plotting

```python
# select a random index from the list of files
curr_idx = np.random.randint(0, len(corpus_files))

# process the text file at the selected index
corpus = Counting(corpus_files[curr_idx])

# plot the Zipf's Plot
corpus.plot_zipf_law()

# plot the Heap's Plot
corpus.heaps_law()

del corpus
```

```python
for curr_idx in range(len(corpus_files)):
    # process the text file at the selected index
    corpus = Counting(corpus_files[curr_idx])

    # plot the Zipf's Plot
    corpus.plot_zipf_law()

    del corpus
```
